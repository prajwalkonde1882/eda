# Different graphs and use cases

* Scatter plot = to understand relationship between two variable
* Line chart = whenever on x axis, we have time component
* Bar chart = for categorical data
* Count plot = for categorical data
* Pie chart = for the distribution of different unique classes
* Box plot and violin plot = To determine the outlier
* Histogram/distribution = for continuous data
* Heatmap = to understand the correlation
* Pairplot = all the numerical relationship in one plot
* Join plot = relationship and distribution of area
* swarmplot/replot = scatterplot
* lmplot = reg line

```python
def clean_reviews(review_str):
    if isinstance(review_str, str):
        if 'M' in review_str:
            return float(review_str.replace('M', '')) * 1_000_000
    return review_str

df_copy['Reviews'] = df_copy['Reviews'].apply(clean_reviews)
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

# Replace string values

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('+', '')
```

```python
df_copy['Size'] = df_copy['Size'].str.replace(',', '')
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('Varies with device', str(np.nan))
```

```python
df_copy['Size'].astype(float)
```

```python
df_copy['Installs']
```

```python
df_copy['Installs'].str.replace('+', '').str.replace(',' , '')
```

```python
df_copy['Installs'].unique()
```

```python
# Remove 'Free' string from Installs column
df_copy = df_copy[df_copy['Installs'] != 'Free'].copy()

# Remove '+' and ',' characters and then convert to integer
df_copy['Installs'] = df_copy['Installs'].astype(str).str.replace('+', '', regex=False).str.replace(',', '', regex=False).astype(int)
```
