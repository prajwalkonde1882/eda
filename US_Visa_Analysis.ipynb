```python
df.drop('case_no', axis = 1, inplace = True)
```

```python
df = df[df['case_status'] != 'Withdrawn']  # withdrawn application should not be analysed
```

```python
# change certified-expired to only certified
df.loc[df.case_status == 'Certified-Expired', 'case_status'] = 'Certified'
```

```python
df.dropna(axis = 1 , how = 'all', inplace = True) # axis 1 means all the columns,
#all means if all values in columns are null then it will droped
```

```python
df.dropna(axis = 0 , how = 'all', inplace = True) # axis 0 means rows,
# if all the values in single row are null then the row will deleted
```

```python
# insight for above chart
df['decision_date'] = pd.to_datetime(df['decision_date'])
df['year'] = df['decision_date'].dt.year
df['year']
```

```python
fig, ax = plt.subplots(figsize = (8,8))
sns.countplot(x = 'year', hue = 'case_status', data = df)
ax.set(xlabel = 'Visa application year', ylabel = 'No. of applications')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,8))

sns.countplot(x = 'employer_city', hue = 'year', data = df, order = df.employer_city.value_counts().iloc[:10].index)
ax.set(xlabel = 'Visa application year', ylabel = 'no. of applications')
plt.xticks(rotation = 90)
plt.show()
```
